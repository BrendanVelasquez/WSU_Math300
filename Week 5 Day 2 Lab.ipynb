### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import sympy as sp

x_values = np.linspace(-10,5,100)
y_values = x_values**2 + 6*x_values+10

plt.plot(x_values,y_values)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = x^2 + 6x + 10")
plt.grid()
plt.show()

def f(x):
    return x[0]**2 + 6*x[0] + 10

result = minimize(f, x0=[0])
print(result.x)
print(result.fun)

# minimum is around x = -3 and the minimum value is around 1
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
x_values = np.linspace(0,50,100)
profit = 50*x_values - x_values**2

plt.plot(x_values, profit)
plt.xlabel("x")
plt.ylabel("Profit")
plt.title("Profit = 50x - x^2")
plt.grid()
plt.show()

def negative_profit(x):
    return -(50*x[0] - x[0]**2)

result = minimize(negative_profit, x0=[10])
best_x = result.x[0]
max_profit = 50*best_x - best_x**2
print(best_x)
print(max_profit)

# maximum happens around x = 25 and maximum value is about 625
````

## Problem 3

Minimize

```python
np.sin(x)+x**2
```

using different starting values.

Do you obtain the same solution? Compute the derivative using `sympy`. When is this derivative zero?

````python
def h(x):
    return np.sin(x[0]) + x[0]**2
for start in [-5,0,5]:
    result = minimize(h,x0=[start])
    print("start:",start)
    print("x:",result.x)
    print("value:",result.fun)

x = sp.symbols("x")
expr = sp.sin(x) + x**2

derivative = sp.diff(expr, x)
print(derivative)
critical_point = sp.nsolve(derivative, -1)
print(critical_point)

# derivative is cos(x) + 2x, the derivative is zero around x = -0.45
# and scipy gives about the same answer from different starting values
````

### Problem 4
Design a function with multiple minima and optimize.

````python
def g(x):
    return np.sin(x[0])+0.1*x[0]**2

x_values = np.linspace(-10,10,300)
y_values = np.sin(x_values) +0.1*x_values**2
plt.plot(x_values,y_values)
plt.xlabel("x")
plt.ylabel("g(x)")
plt.title("Function With Multiple Minima")
plt.grid()
plt.show()

for start in [-8,-2,2, 8]:
    result = minimize(g, x0=[start])
    print("start:",start)
    print("x:",result.x)
    print("value:",result.fun)

# this one has multiple dips, so depending on the starting value, scipy can find different local minimums
````
