```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from itertools import combinations

#problem 1
print("\nProblem 1")
iris=pd.read_csv("iris.csv")
print(iris.columns)
print(iris.head())
plt.figure(figsize=(7,5))
plt.scatter(iris["sepal.length"],iris["sepal.width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length Compared to Sepal Width")
plt.show()

# There are a few points separated from the rest,but no obvious outliers
#the relationship looks weak because the points are fairly spread out

variety_counts=iris["variety"].value_counts()
print("\nvariety counts:")
print(variety_counts)
plt.figure(figsize=(7,5))
plt.bar(variety_counts.index,variety_counts.values)
plt.xlabel("Flower variety")
plt.ylabel("Frequency")
plt.title("Frequency of iris varieties")
plt.show()
print("most common variety is:")
print(variety_counts[variety_counts == variety_counts.max()].index.tolist())
# all three varieties are tied at 50

#problem 2
print("\nProblem 2")
dolphins=pd.read_csv("DolphinNetwork.csv")
dolphin_graph=nx.from_pandas_edgelist(
    dolphins,
    source="Source",
    target="Target")
plt.figure(figsize=(11,8))
positions=nx.spring_layout(dolphin_graph,seed=3)
nx.draw(
    dolphin_graph,
    positions,
    node_color="lightblue",
    edge_color="gray",
    node_size=450,
    with_labels=True,
    font_size=7)
plt.title("Dolphin social Network")
plt.show()

def count_dolphin_triangles(G):
    counts ={}
    for dolphin in G.nodes():
        neighbors = list(G.neighbors(dolphin))
        triangle_count = 0
        for dolphin1, dolphin2 in combinations(neighbors,2):
            if G.has_edge(dolphin1,dolphin2):
                triangle_count +=1
        counts[dolphin] = triangle_count
    return counts
    
triangle_counts=count_dolphin_triangles(dolphin_graph)
triangle_data=pd.Series(triangle_counts).sort_index()
print("triangle counts:")
print(triangle_data)
plt.figure(figsize=(14,6))
plt.bar(triangle_data.index,triangle_data.values)
plt.xlabel("Dolphin")
plt.ylabel("Number of Triangles")
plt.title("Triangle Count for Each Dolphin")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

highest_count=triangle_data.max()
lowest_count=triangle_data.min()
print("highest triangle count:",highest_count)
print("dolphin or dolphins with highest count:")
print(triangle_data[triangle_data==highest_count].index.tolist())
print("lowest triangle count:",lowest_count)
print("dolphin or dolphins with lowest count:")
print(triangle_data[triangle_data==lowest_count].index.tolist())
print("mean triangle count:",triangle_data.mean())
```
