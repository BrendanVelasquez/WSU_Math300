### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-1,1,10)
x2 = np.linspace(-1,1,100)
x3 = np.linspace(-1,1,1000)

figure, axes = plt.subplots(1,3,figsize=(12,4),sharey=True)

axes[0].plot(x1, x1**2)
axes[0].set_title("10 slices")

axes[1].plot(x2, x2**2)
axes[1].set_title("100 slices")

axes[2].plot(x3, x3**2)
axes[2].set_title("1000 slices")

figure.suptitle("Plot of x^2")
```

### Problem 4
Download the "WA_County_Boundaries.shp" file. 
- Plot the state using `name_of_file_read.plot()`
- Plot some information about the state by extracting two columns of data.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
wa = gpd.read_file("WA_County_Boundaries.shp")

#first plot for the map
wa.plot()
plt.axis("off")
plt.title("Washington County Boundaries")
plt.show()

# check columns for data
print(wa.columns)

#make two simple columns
wa["area"] = wa.geometry.area
wa["border_length"] = wa.geometry.length

#now make the second plot: scatter
plt.figure()
plt.scatter(wa["area"], wa["border_length"])
plt.xlabel("Area")
plt.ylabel("Length")
plt.title("County Area vs Border Length")
plt.show()
```
