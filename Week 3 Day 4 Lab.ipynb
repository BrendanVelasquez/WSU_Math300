### Problem 1
We are going to make a social network. This can be real or made up. Up to you!
1. Add nodes representing people
2. Add edges if these people talk
3. Make the edges bigger if they are good friends and smaller if they are just acquaintances
4. Make the nodes size correspond to height
5. Make the node color correspond to hair color (i.e. if hair is brunette, the node is brown)


Make sure your final plot has:
- a title
- the names of the people displayed
- color
- is legible and presentable
- an appropriate sizing

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

friends = nx.Graph()
# these are people that talk to each other
friends.add_edges_from([
    ("Quinton","Shawn"), ("Quinton", "Carlos"), ("Shawn","Carlos"),
    ("Shawn", "Dylan"), ("Carlos","Emma"), ("Emma", "Faith"), ("Dylan","Faith")
])

#taller people have larger circles, the bigger the number the larger the circle the taller.
sizes = [1400, 1850,1600, 2400, 1550,1750]
#colors are supposed to be their hair color
colors = ["brown","gold","white", "saddlebrown","orange","silver"]
# these have to be in the same order as the edges
# bigger number means they are better friends, lower means worse
edgeWidths = [5,2, 4,1,3, 2,1]

plt.figure(figsize=(9,7))
nx.draw(friends,
        with_labels = True,
        node_size=sizes,
        node_color = colors,
        width=edgeWidths,
        edge_color="gray",
        edgecolors="black",
        linewidths=2,
        font_size = 11)
plt.title("Friend Social Network")
plt.show()
```

### Problem 2

For the dolphin social network:
1. Import the data
2. Draw the network 
3. Use as many `networkx` parameters as possible (that is, in `nx.draw()`, add all the parameters: `node_size`, `node_color`, etc.) such that the network is still presentable and understandable

You can find some information about `nx.draw()` here: https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw.html#networkx.drawing.nx_pylab.draw .


```python
# load the data for the dolphin network problem
dolphins = pd.read_csv("DolphinNetwork.csv")
dolphins.head()
edges = []

#turn each row into a pair for networkx
for i in range(len(dolphins)):
    pair = (dolphins["Source"].iloc[i],dolphins["Target"].iloc[i])
    edges.append(pair)

dolphinGraph = nx.Graph()
dolphinGraph.add_edges_from(edges)

plt.figure(figsize = (14,14))
nx.draw(dolphinGraph,
        with_labels=True,
        node_size = 500,
        node_color="lightblue",
        edge_color = "gray",
        width=1.5,
        edgecolors="darkblue",
        linewidths = 1,
        font_size=7,
        font_color="black",
        alpha=.85)
plt.title("Dolphin Social Network",fontsize=18)
plt.show()
```
