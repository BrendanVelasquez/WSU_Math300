### Problem 4: Summary Statistics
For each numerical column, compute the following:
- Mean
- Median
- Mode
- Minimum
- Maximum
- Standard deviation


1. Which variable has the greatest variation?
2. Which variable has the smallest variation?


````python
print("mean")
print(df.mean(numeric_only=True))
# median gives the middle for each number column which this is useful because it is less affected by extreme values than the mean
print("median")
print(df.median(numeric_only=True))

# mode gives the most common value
print("mode")
print(df.mode())

# min gives the smallest value in each number column, max is biggest
print("min")
print(df.min(numeric_only=True))

print("max")
print(df.max(numeric_only=True))

#bigger standard deviation means more variation
print("standard deviation")
stds = df.std(numeric_only=True)
print(stds)

# checking variation
# From the output it can br seen that petal.length has the greatest variation because it has the biggest standard deviation
print("greatest variation")
print(stds.idxmax())

# sepal.width has the smallest variation because it has the smallest standard deviation
print("smallest variation")
print(stds.idxmin())
````

### Problem 5: Visualization
1. Choose a numerical column from your dataset and:
    - Plot the corresponding histogram. 
        - What does the distribution look like?
        - Is it symmetric or skewed?
    - Plot the corresponding box plot.
        - Are there any outliers?
        - Explain what you observe.
2. Choose one categorical column from your dataset and:
    - Plot the corresponding bar chart.
        - Which category occurs most frequently?
        - Which occurs least frequently?


````python
# use petal.length, not petal_length, because the column name has a dot in it
plt.hist(df["petal.length"])
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Length")
plt.show()

# the histogram is not really symmetric
# it looks like there are clusters, probably because the iris types have different petal lengths


# box plot

plt.boxplot(df["petal.length"])
plt.ylabel("Petal Length")
plt.title("Boxplot of Petal Length")
plt.show()

# the boxplot shows the spread of petal length and there are no obvious outliers showing up here


# bar chart
df["variety"].value_counts().plot(kind="bar")

plt.xlabel("Variety")
plt.ylabel("Count")
plt.title("Iris Variety Counts")
plt.show()

# all three iris varieties have the same count
# so there is not really a most common or least common variety
````
