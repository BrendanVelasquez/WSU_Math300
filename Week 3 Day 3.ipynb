### Problem 1
1. Create your own dataframe with at least 3 columns and 4 rows.
2. Save the dataframe to a csv file.
3. Read the csv back in.
4. Plot the data.

````python
import pandas as pd
import matplotlib.pyplot as plt

#part 1
names = ["Shawn", "Quinton", "Jared", "Callie"]
ages = [30, 27, 44, 41]
scores = [85, 90, 78, 88]

df = pd.DataFrame({
    "Name": names,
    "Age": ages,
    "Score": scores
})
print(df)

#part 2
df.to_csv("mydata.csv", index=False)

# part 3
newdf = pd.read_csv("mydata.csv")
newdf
#part 4 
plt.scatter(newdf["Age"], newdf["Score"])
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score")
````

### Problem 3
Download the "WA_County_Boundaries" folder from the "Datasets" folder on GitHub.
- Read the .shp file into Jupyter by using `variable_name = geopandas.read_file("path_to_.shp_file.shp")`
- Plot the state using `variable_name.plot()`
- Plot some information about the state by extracting two columns of data.

````python
# problem 3

import geopandas as gpd
import matplotlib.pyplot as plt

wa = gpd.read_file("WA_County_Boundaries.shp")

# map plot
wa.plot()
plt.axis("off")
plt.title("Washington County Boundaries")
plt.show()

#checking columns
print(wa.columns)

#  make columns
wa["area"] = wa.geometry.area
wa["border_length"] = wa.geometry.length

# the plot
plt.figure()
plt.scatter(wa["area"], wa["border_length"])
plt.xlabel("Area")
plt.ylabel("Border Length")
plt.title("Area vs Border Length")
plt.show()
````
