## Problem 2

1. Write a recursive function `factorial()` that takes in a natural number `n` and computes $n!=n(n-1)(n-2)\cdots3\cdot2\cdot 1$. 
2. Use `%timeit` to determine how fast your function calculates `factorial(150)`. 
3. Use `%timeit` to calculate how fast `math.factorial()` calculates the same quantity.
4. Were there any differences between the two methods?

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)
factorial(5)

# timing my version
%timeit factorial(150)

# timing the built in math version
%timeit math.factorial(150)

#The math factorial one is faster.
```

## Problem 3
Recall for a quadratic equation of the form
$$ax^2 + bx + c = 0
$$

The solutions are given by the <i>quadratic formula</i>:

$$
x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}
$$

1. Write a function `quadratic()` that calculates the two roots given by the quadratic formula only if they are real. 
2. Use `%timeit` to determine how fast your function calculates the roots for $437x^2 - 915x + 478 = 0$. 
3. Use `%timeit` to calculate how fast `sp.solve()` calculates the same quantities.
4. Were there any differences between the two methods?

```python
#quadratic formula function

def quadratic(a,b,c):
    disc = b**2-4*a*c
    if disc<0:
        return "no real roots"
    else:
        root1 = (-b+math.sqrt(disc))/(2*a)
        root2 = (-b-math.sqrt(disc))/(2*a)
        return [root1,root2]
quadratic(437,-915,478)


#timing my function
%timeit quadratic(437,-915,478)

# sympy solve version and timing
x = sp.symbols('x')
sp.solve(437*x**2-915*x+478,x)
%timeit sp.solve(437*x**2 - 915*x + 478,x)

#The biggest differences is that Sympy is slower, and my function will give decimals 
```

## Problem 4

For each of the following functions, use `sympy` calculate the first derivative, the second derivative, and the integral (all with respect to $x$):

1. $f(x)=x$
2. $f(x)=6x^{-1/2}$
3. $f(x)=\pi\sqrt{x}+x^5+7$
4. $f(x)=x(2x-5)(3+\tan(x))$

```python
# derivatives and integrals with sympy

x = sp.symbols('x')

print("Problem 1")
#1. f(x) = x
f = x
print("first derivative:")
print(sp.diff(f,x))
print("second derivative:")
print(sp.diff(f,x,2))
print("integral:")
print(sp.integrate(f,x))
print(" ")

print("Problem 2")
#2. f(x) = 6x^(-1/2)
f = 6*x**(-1/2)
print("first derivative:")
print(sp.diff(f,x))
print("second derivative:")
print(sp.diff(f,x,2))
print("integral:")
print(sp.integrate(f,x))
print(" ")

print("Problem 3")
#3. f(x) = pi*sqrt(x)+x^5 + 7
f = sp.pi*sp.sqrt(x) + x**5+ 7
print("first derivative:")
print(sp.diff(f,x))
print("second derivative:")
print(sp.diff(f,x,2))
print("integral:")
print(sp.integrate(f,x))
print(" ")

print("Problem 4")
# 4. f(x) =x(2x-5)(3+tan(x))
f = x*(2*x-5)*(3+sp.tan(x))
print("first derivative:")
print(sp.diff(f,x))
print("second derivative:")
print(sp.diff(f,x,2))
print("integral:")
print(sp.integrate(f,x))

```

## Problem 5

From yesterday, we calculated <i>numerically</i> what the derivative of $\sin(x)$ is. 
1. Using sympy, calculate derivative at 0.
2. Using sympy, evaluate the limit definition of the derivative for $f(x)=\sin(x)$:
$$f'(x)=\lim_{h\to0}\frac{f(x+h)-f(x)}{h}
$$

```python
#derivative of sin(x) using sympy

x = sp.symbols('x')
f = sp.sin(x)
deriv = sp.diff(f,x)
print(deriv)
print(deriv.subs(x,0))
```

```python
# limit definition and at x = 0

x,h = sp.symbols('x h')
limitThing = (sp.sin(x+h) - sp.sin(x))/h
print(sp.limit(limitThing,h,0))
print(sp.limit(limitThing,h,0).subs(x,0))
```
